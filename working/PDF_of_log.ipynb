### Pdf of the log
1. Use `numpy` to draw $N$ (where $N$ is large...) random samples from a uniform distribution between $0.1$ and $10$, and store these samples as $x$.

```python
import numpy as np
```

```python
x = np.random.uniform(0.1, 10, 10000)
```

2. Use matplotlib to make a histogram of these samples.

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x, 100, histtype='barstacked', density=True, alpha=0.4, edgecolor='none', label='x distribution');
plt.ylabel('Probability density')
plt.xlabel('X range');
```

3. Compute the base-10 log of your array $x$, and store this as $y$.

```python
y = np.log10(x)
```

4. Make another histogram for $y$. Using the equation to transform probability distributions, write what the theoretical pdf of $y$ is, and overplot it onto your histogram.

```python
plt.hist(y, 100, histtype='barstacked', density=True, alpha=0.4, edgecolor='none', label='y distribution');
plt.legend();
plt.ylabel('Probability density')
plt.xlabel('Y range');
```

The equation is: $p(y) = |\frac{dx}{dy}|p(x)$.

In my case: 

1. $p(x) = \frac{1}{10 - 0.1} = \frac{1}{9.9}$; 
2. $y = \log{x}$, meaning that $x = 10^{y}$: this means that $|\frac{dx}{dy}| = 10^{y} \cdot \ln{10}$

3. $p(y) = \frac{10^{y} \cdot \ln{10}}{9.9}$   

```python
z = np.linspace(-1, 1, 10000)
p_y = []

for i in z: 

    p_y.append(10**i * np.log(10)/ 9.9)
```

```python
plt.figure(figsize=(10, 6))

plt.hist(y, 100, histtype='barstacked', density=True, alpha=0.4, edgecolor='none', label='y distribution');
plt.plot(z, p_y, color='C0', ls='--', label='theoretical p(y)')

plt.legend()
plt.ylabel('Probability density')
plt.xlabel('Y range');
```

5. Compute the log of the mean of $x$ and the mean of $y$. Now compute the log of the median of $x$ and the median of $y$. 

```python
log_mean_x = np.log10(np.mean(x))
log_mean_y = np.mean(y)

print(log_mean_x, log_mean_y)
```

```python
log_median_x = np.log10(np.median(x))
log_median_y = np.median(y)

print(log_median_x, log_median_y)
```

As explained the median does not change, since it only depends on the ordering of values. The mean changes instead because it depends on the scale of the distribution.
