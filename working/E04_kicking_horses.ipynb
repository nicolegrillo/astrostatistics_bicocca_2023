### 2. Kicking horses

A famous early  application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by [Bortkiewicz](https://www.wikiwand.com/en/Ladislaus_Bortkiewicz) in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

- Plot the resulting probability distribution. Careful with the normalization
- How does it look like? Again try to fit it by hand.

(Hint: think about sample mean and sample variance)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

First I construct the distribution: 

```python
dist_men = np.array([0] * 109 + [1] * 65 + [2] * 22 + [3] * 3 + [4] * 1)
men = [0, 1, 2, 3, 4]
```

Then I plot it: 

```python
plt.figure(figsize = (10, 6.5))
plt.hist(dist_men, color = 'lightblue', label = '#Deaths distribution vs #Groups');  # is this the correct way to normalize?

plt.xlabel('Deaths')
plt.ylabel('Groups')
plt.legend();
```

Let's suppose data could be described by the Poisson distribution, which is applied to small discrete samples of data. I will define it as it follows: 

```python
x_i = np.linspace(0, 4, 5)

mean = np.mean(dist_men)
poisson_distr = poisson.pmf(dist_men, mu = mean)
```

Plotting the results: I first need to normalize the results knowing that they are discrete. This means that I need to use the simplest definition of probability: 

$P(x_i) = \frac{x_i}{N}$

where $\textit{N}$ is the total number of samples $\textit{x_i}$. 

```python
dist_men_normalized = [np.count_nonzero(dist_men == 0) / len(dist_men), np.count_nonzero(dist_men == 1) / len(dist_men), np.count_nonzero(dist_men == 2) / len(dist_men), np.count_nonzero(dist_men == 3) / len(dist_men), np.count_nonzero(dist_men == 4) / len(dist_men)]
```

```python
plt.figure(figsize = (10, 6.5))
plt.bar(men, dist_men_normalized, color = 'lightblue', label = '#Groups VS #Deaths distribution');  # is this the correct way to normalize?
plt.plot(dist_men, poisson_distr , color='C1', ls='--', label='Poisson distribution')

plt.xlabel('Groups')
plt.ylabel('Deaths')
plt.legend();
```

In this last plot I confirmed that the distribution could be considered as a Poissonian distribution.
